# Insertar fondo transparente a imagen

## Parte 1, cambiar a blanco el color de fondo

Importamos las librer√≠as

```python
import numpy as np
import imageio
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

Abrimos la imagen y capturamos su altura y anchura

```python
nombre_fichero = cv2.imread("image.png")

h, w = nombre_fichero.shape[:2]
```

Mostramos por pantalla la imagen, debido a que plt utiliza los filtros RGB y cv2 BGR debemos realizar un cambio de orden.

```python
plt.imshow( cv2.cvtColor(nombre_fichero, cv2.COLOR_BGR2RGBA) )
print("Estructura: ",nombre_fichero.shape)
```

Creamos una matriz de ceros dos p√≠xeles m√°s grande en altura y anchura

```python
mask = np.zeros((h+2, w+2), dtype = np.uint8)
```

El m√©todo `.floodFill()` rellena un objeto conectado con un color dado.
-  El primer argumento especifica la entrada, la imagen a rellenar
-  El segundo argumento, `mask` debe ser una imagen de 8 bits de un solo canal, 2 p√≠xeles mas ancha y alta que la imagen.
-  El p√≠xel inicial, el color de ese p√≠xel ser√° el que conectaremos con los dem√°s p√≠xeles
-  El color del p√≠xel nuevo
-  M√°xima diferencia __inferior__ de color entre el p√≠xel observado y el de uno de sus vecinos que pertenezcan al componente
-  M√°xima diferencia __superior__ de color entre el p√≠xel observado y el de uno de sus vecinos que pertenezcan al componente.

```python
dif_valor = 6  #Este valor es distinto para cada imagen
diff = (dif_valor, dif_valor, dif_valor)
```

```python
cv2.floodFill(nombre_fichero, mask, (200,10), (255, 255, 255), diff, diff)

print(nombre_fichero.shape)
plt.imshow(cv2.cvtColor(nombre_fichero, cv2.COLOR_BGR2RGB))
plt.figure()
```

## Parte 2.1, trasformar fondo blanco a fondo transparente

A partir de una imagen con fondo blanco transformamos los p√≠xeles blancos en transparentes mediante la librer√≠a Image, transformamos nuestra imagen de tipo cv2 a tipo Image

```python
img = Image.fromarray(cv2.cvtColor(nombre_fichero, cv2.COLOR_BGR2RGB))
```

Transformamos la imagen RGB a RGBA, la variable A define la opacidad del p√≠xel, por lo que para ciertos p√≠xeles insertaremos opacidad 0.

```python
img = img.convert("RGBA")
```

```python
datas = img.getdata()
```

Creamos variable newData donde insertaremos los p√≠xeles de la nueva imagen

```python
newData = []
```

Recorremos todos los p√≠xeles de la imagen, donde el valor del p√≠xel sea totalmente blanco (255, 255, 255) ponemos intensidad 0.

```python
for item in datas:
    if item[0] == 255 and item[1] == 255 and item[2] == 255:
        newData.append((255,255,255,0))
    else:
        newData.append(item)
```

Guardamos la imagen

```python
img.putdata(newData)
img.save("transparentBackgroundImage.png", "PNG")
```

## Parte 2.2 Otra forma transformar fondo blanco a transparente

Transformamos la imagen a una matriz mediante la librer√≠a `numpy`

```python
imagen1 = np.array(Image.fromarray(cv2.cvtColor(nombre_fichero, cv2.COLOR_BGR2RGBA)))
```

```python
altura, anchura = imagen1.shape[:2]
```

Realizamos una copia de nuestra imagen actual

```python
imagenSinFondo = imagen1
```

Si encontramos un p√≠xel de color blanco total (255, 255, 255) le asignamos opacidad 0.

```python
for i in range(altura):
    for j in range(anchura):
        if (imagenSinFondo[i,j,0] == 255 and imagenSinFondo[i,j,1] == 255 and imagenSinFondo[i,j,2] == 255 ):
            imagenSinFondo[i,j,3] = 0
```

Guardamos la imagen mediante la librer√≠a `imageio`

```python
imageio.imwrite('transparentBackgroundImage2.png', imagenSinFondo)
```
